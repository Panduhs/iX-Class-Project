### Aggregating the number of tickets sold for each ride

```python
tickets_df = df1.groupby(by=['ride_id'])['ride_id'].count().to_frame()
tickets_df = tickets_df.rename(columns = {'ride_id':'tickets'})
df1 = pd.merge(df1,tickets_df,on='ride_id')
df1.head()
```

## 2.2 Converting object dtypes to numerical values

### Splitting the travel_date to year, month and date

```python
df1['year'] = pd.DatetimeIndex(df1['travel_date']).year
df1['month'] = pd.DatetimeIndex(df1['travel_date']).month
df1['day'] = pd.DatetimeIndex(df1['travel_date']).day
df1.dtypes
```

### Converting travel_date to weekday (0 is monday and so on)

```python
df1['travel_date'] = pd.to_datetime(df1["travel_date"], format = "%d-%m-%y")
df1["weekday"] = df1["travel_date"].dt.dayofweek
df1.dtypes
```

### Converting travel_time to minutes from 00:00 and getting the departure hour

```python
df1["departure_mins"] = df1["travel_time"].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
df1['hour'] = df1.travel_time.map(lambda x: pd.to_datetime(x).hour)
```

```python
df1.head()
```

### Converting travel_from to corresponding integer

```python
df1["travel_from"] = pd.Categorical(df1["travel_from"])
travel_from_categories = df1.travel_from.cat.categories
df1["travel_from_cat"] = df1.travel_from.cat.codes
df1["travel_from_cat"] = df1["travel_from_cat"] + 1
```

### Cleaning up the data set

```python
cols_to_drop = ['ride_id', 'travel_date', 'travel_time', 
                'travel_to']
df2 = df1.drop(columns = cols_to_drop, axis=0)
df2.head()
```
